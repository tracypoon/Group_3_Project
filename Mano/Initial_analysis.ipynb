```python
# Dependencies
import pandas as pd
import calendar
import matplotlib.pyplot as plt

import scipy.stats as st
```

```python
# Name of the CSV file
file = '../Mano/Final_Covid_Data.csv'
```

```python
df = pd.read_csv(file)
df
```

```python
df['Month'] = df['Month'].apply(lambda x: calendar.month_abbr[x])
```

```python
df
```

```python
df=df.drop(columns='Unnamed: 0')
```

```python
df
```

```python
df.describe()
```

```python
#Calculating the actual case count from the running total
total_case=0
current_cases = 0
previous_cases = 0 
first_time = 1
months = []
day=[]
cases = []


for index,row in df.iterrows():
#     print (row["Month"])
#     break
    
    total_case = int(row['Total_Cases'])
#     print(total_cases)
    
    if first_time == 1:
        first_time = 0
    else:    
#         print(f"previous= {previous_cases}")
    
        current_cases = total_case - previous_cases
        day.append(row['Day'])
        months.append(row['Month'])
        cases.append(current_cases) 
        #print(f"previous= {previous_cases}")
        #print(f"month= {month_data}")
        
    previous_cases = total_case
    
month_data_df = pd.DataFrame({
    "Month": months,
    "Day": day,
    "Total_cases": cases
})

month_data_df
```

```python
df1=month_data_df.groupby('Month').agg({"Total_cases": "sum"})
ordered= ['Jan','Feb', 'Mar', 'Apr','May', 'Jun', 'Jul', 'Aug', 'Sep','Oct', 'Nov']
df2=df1.reindex(ordered)
df2
#df3=df2.reset_index()
#df3
```
